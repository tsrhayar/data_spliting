```python
import csv

def lire_fichier_csv(nom_fichier):
    lignes = []
    with open(nom_fichier, 'r', newline='') as fichier_csv:
        lecteur_csv = csv.reader(fichier_csv)
        for ligne in lecteur_csv:
            lignes.append(ligne)
    return lignes

def diviser_fichier_csv(nom_fichier, pourcentage_partie1, pourcentage_partie2):
    global df1, df2, df3
    lignes = lire_fichier_csv(nom_fichier)
    entetes = lignes[0]  # Première ligne contenant les noms de colonnes
    donnees = lignes[1:]  # Le reste des données sans la première ligne

    taille_fichier = len(donnees)

    taille_partie1 = int(taille_fichier * pourcentage_partie1 / 100)
    taille_partie2 = int(taille_fichier * pourcentage_partie2 / 100)
    taille_partie3 = taille_fichier - taille_partie1 - taille_partie2

    df1 = [entetes] + donnees[:taille_partie1]
    df2 = [entetes] + donnees[taille_partie1:taille_partie1+taille_partie2]
    df3 = [entetes] + donnees[taille_partie1+taille_partie2:]

    ecrire_fichier_csv('part1.csv', df1)

def ecrire_fichier_csv(nom_fichier, lignes):
    with open(nom_fichier, 'w', newline='') as fichier_csv:
        ecrivain_csv = csv.writer(fichier_csv)
        ecrivain_csv.writerows(lignes)

# Exemple d'utilisation : Diviser le fichier en 50%, 30%, 20%
nom_fichier_csv = 'dataset_olympics.csv'
pourcentage_partie1 = 10
pourcentage_partie2 = 5

diviser_fichier_csv(nom_fichier_csv, pourcentage_partie1, pourcentage_partie2)
```

```python
import mysql.connector

# Replace with your actual MySQL credentials and database information
db_config = {
    "user": "root",
    "password": "",
    "host": "localhost",
    "database": "olympic"
}


try:
    # Establish a connection to the MySQL database
    connection = mysql.connector.connect(**db_config)
    cursor = connection.cursor()
    
    # Drop the table if it already exists
    drop_table_query = "DROP TABLE IF EXISTS olympic"
    cursor.execute(drop_table_query)

    # Create the table if it doesn't exist
    create_table_query = """
    CREATE TABLE IF NOT EXISTS olympic (
        ID INT PRIMARY KEY AUTO_INCREMENT,
        Name VARCHAR(255),
        Sex VARCHAR(1),
        Age FLOAT,
        Height FLOAT,
        Weight FLOAT,
        Team VARCHAR(255),
        NOC VARCHAR(3),
        Games VARCHAR(255),
        Year INT,
        Season VARCHAR(10),
        City VARCHAR(255),
        Sport VARCHAR(255),
        Event VARCHAR(255),
        Medal VARCHAR(20)
    )
    """
    cursor.execute(create_table_query)

    # Convert data to a pandas DataFrame
    df3 = pd.DataFrame(df3[1:], columns=df3[0])
    
    # Replace empty strings in the "Age" column with None (NULL)
    df3['Age'] = df3['Age'].replace('', None)
    
    # Replace empty strings in the "Age" column with None (NULL)
    df3['Height'] = df3['Height'].replace('', None)
    
    # Replace empty strings in the "Age" column with None (NULL)
    df3['Weight'] = df3['Weight'].replace('', None)
    
    # Convert DataFrame to a list of tuples (one tuple per row)
    rows = [tuple(row) for _, row in df3.iterrows()]

    
    
    # SQL query for bulk insert, excluding the ID column
    insert_query = "INSERT INTO olympic (Name, Sex, Age, Height, Weight, Team, NOC, Games, Year, Season, City, Sport, Event, Medal) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"

    # Use executemany to perform bulk insert
    cursor.executemany(insert_query, rows)

    # Commit the changes
    connection.commit()

except Exception as e:
    # Rollback the changes in case of any error
    connection.rollback()
    print("Error occurred during data insertion:", e)

finally:
    # Close the connection
    if cursor:
        cursor.close()
    if connection:
        connection.close()
        
        print('Done')
```
